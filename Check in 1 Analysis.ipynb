# Semester 2 Check in 1 Analysis

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
issues_df = pd.read_csv("Issues_feb.csv")
```

```python
issues_df
```

```python
issues_df['Delta'] = issues_df['Actual Estimate'] - issues_df['Original Estimate']
```

```python
issues_df[ 'Delta'].hist()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(10, 8))
#plot the number of people 
data = issues_df["Delta"].to_numpy()

N, bins, patches = ax.hist(data, edgecolor='white', bins=8, linewidth=1)


# patches[0].set_facecolor('red')
# patches[1].set_facecolor('orange')
# patches[2].set_facecolor('blue')
# patches[3].set_facecolor('cyan')
# patches[4].set_facecolor('magenta')
# patches[5].set_facecolor('#5599ff')

for i in range(0, len(patches)):
    patches[i].set_facecolor('magenta')

plt.xlim([-2, 7])
plt.xticks(np.arange(-2, 8, 1.0))

plt.xlabel("Delta")
plt.ylabel("Number of Issues")
plt.grid(axis='y')
plt.title(label="Histogram of Issue Deltas")
# bins_labels(bins, fontsize=20)

plt.show()
```

```python
groups = issues_df.groupby('Asignee') 
```

```python
for name, group in groups:
    fig, ax = plt.subplots(figsize=(10, 10))
    group["Delta"].hist()
    plt.xlabel("Delta")
    plt.ylabel("Number of Issues")
    plt.grid(axis="x")
    plt.title(label="Histogram of Issue Deltas for Developer: {}".format(name))
    
    plt.show()
```

## Time Series Data


```python
issues_df['Date Assigned'] = pd.to_datetime(issues_df['Date Assigned'], infer_datetime_format=True)  
```

```python
issues_df = issues_df.sort_values(by="Date Assigned")
```

```python
plt.plot(issues_df['Date Assigned'],issues_df['Delta'], linestyle='--', marker='o', color='b')
plt.title(label="Delta over time for Developers")
plt.xlabel("Date")
plt.ylabel("Delta")
plt.xticks(rotation=90)
plt.grid(axis="y")
plt.show()
```

### Group by Developer Time Series Data


```python
for name, group in groups:
    group = group.sort_values(by="Date Assigned")
    fig, ax = plt.subplots(figsize=(10, 10))
    plt.plot(group['Date Assigned'],group['Delta'], linestyle='--', marker='o', color='b')
    plt.title(label="Delta over time for Developer: {}".format(name))
    plt.xlabel("Date")
    plt.ylabel("Delta")
    plt.xticks(rotation=90)
    plt.grid(axis="y")
    plt.show()
```
